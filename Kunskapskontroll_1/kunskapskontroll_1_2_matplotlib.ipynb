### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

There are two ways. One is the object-oriented style: explicitly creating Figures and Axes and calling methods on them. The other is PyPlot style, creating Figures and Axes implicitly and using PyPlot functions on them. 
The object-oriented approach is recommended.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

The figure is the entire plot with axes and artists. Axes is the region that contains plotted data, with Axis objects. Axis is the object that set scale, limits, ticks and tick labels. Artists are all objects that are visible on the Figure.

### When plotting in Matplotlib, what is the expected input data type?

The expected input data type is a numpy array.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

Object-oriented approach:

```python
x = np.linspace(-4, 4)
y = x**2

fig, ax = plt.subplots(facecolor='lightblue')
ax.plot(x, y, color='xkcd:blueberry', linewidth=2)
ax.set_facecolor('ivory')
ax.set_xlabel('x', fontsize=14, color='xkcd:darkish blue', fontweight='bold', labelpad=10)
ax.set_ylabel('y', fontsize=14, color='xkcd:darkish blue', fontweight='bold', rotation=0, labelpad=10)
ax.set_title('plot of the function $ y=x^2 $', fontname='Georgia', fontweight='bold', fontsize=16, color='xkcd:darkish blue') 
ax.text(-3.5, 14, r'$y=x^2$', fontsize=15, color='xkcd:blueberry')
ax.grid(True)
plt.show()
```

PyPlot approach:

```python
x = np.linspace(-4, 4)
y = x**2

plt.figure(facecolor='xkcd:pastel pink')
plt.axes(facecolor='xkcd:off white')
plt.plot(x, y, color='xkcd:aubergine', linewidth=2)
plt.xlabel('x', fontsize=14, color='xkcd:chocolate brown', fontweight='bold', labelpad=10)
plt.ylabel('y', fontsize=14, color='xkcd:chocolate brown', fontweight='bold', rotation=0, labelpad=10)
plt.title('plot of the function $ y=x^2 $', fontname='Georgia', fontweight='bold', fontsize=16, color='xkcd:chocolate brown')
plt.text(-3.5, 14, r'$y=x^2$', fontsize=15, color='xkcd:aubergine')
plt.grid(True)
plt.show()
```
